```python
from scipy import misc

IMAGE = 'cat.png'

def dist(x, y):
    return sum((x - y) ** 2)

def flip(i, j, A, B):
    if (i, j) in A:
        A.remove((i, j))
        B.add((i, j))
    else:
        B.remove((i, j))
        A.add((i, j))

def main():
    image = misc.imread(IMAGE)
    dim_x, dim_y, colors = image.shape
    print("Loaded image of shape {x}, {y}".format(x=dim_x, y=dim_y))
    curr1 = [66, 56, 71, 0]
    curr2 = [255, 128, 102, 0]
    last1 = [255, 128, 102, 0]
    last2 = [66, 56, 71, 0]
    
    
    dim_x, dim_y, colors = image.shape
    while (curr1 != last1) or (curr2 != last2):
        A = set()
        B = {(i, j) for i in range(dim_x) for j in range(dim_y)}
        for i in range(dim_x):
            for j in range(dim_y):
                if (dist(image[i, j], curr1) < dist(image[i, j], curr2)):
                    flip(i, j, A, B)
        x = [0, 0, 0, 0]
        for el in A:
            c_a = image[el[0], el[1]]
            x[0] += c_a[0]
            x[1] += c_a[1]
            x[2] += c_a[2]
        x[0] /= len(A)
        x[1] /= len(A)
        x[2] /= len(A)
        y = [1, 1, 1, 1]
        for el in B:
            c_b = image[el[0], el[1]]
            y[0] += c_b[0]
            y[1] += c_b[1]
            y[2] += c_b[2]   
        y[0] /= len(B)
        y[1] /= len(B)
        y[2] /= len(B)
        last1 = curr1
        last2 = curr2
        curr1 = x
        curr2 = y

    for i in range(dim_x):
        for j in range(dim_y):
            if (image[i, j, 3] != 100):
                if (i, j) in A:
                    image[i, j, 0] = 255
                    image[i, j, 1] = 0
                    image[i, j, 2] = 0
                elif (i,j) in B:
                    image[i, j, 0] = 0
                    image[i, j, 1] = 255
                    image[i, j, 2] = 255


    misc.imsave('out_cat.png', image)

if __name__ == '__main__':
    main()
```

```python
from scipy import misc

IMAGE = 'cat2.jpg'

def dist(x, y):
    return sum((x - y) ** 2)

def flip(i, j, A, B):
    if (i, j) in A:
        A.remove((i, j))
        B.add((i, j))
    else:
        B.remove((i, j))
        A.add((i, j))

def main():
    image = misc.imread(IMAGE)
    dim_x, dim_y, colors = image.shape
    print("Loaded image of shape {x}, {y}".format(x=dim_x, y=dim_y))
    
    curr1 = [255, 255, 255]
    curr2 = [255, 128, 102]
    curr3 = [192, 192, 192]
    last1 = [255, 128, 102]
    last2 = [192, 192, 192]
    last3 = [255, 255, 255]
    
    dim_x, dim_y, colors = image.shape
    while (curr1 != last1) or (curr2 != last2) or (curr3 != last3):
        A = set()
        B = {(i, j) for i in range(dim_x) for j in range(dim_y)}
        C = set()
        for i in range(dim_x):
            for j in range(dim_y):
                p = min(dist(image[i, j], curr1), dist(image[i, j], curr2), dist(image[i, j], curr3))
                if (p == dist(image[i, j], curr1)):
                    flip(i, j, A, B)
                elif (p == dist(image[i, j], curr3)):
                    flip(i, j, C, B)
        x = [0, 0, 0]
        for el in A:
            c_a = image[el[0], el[1]]
            x[0] += c_a[0]
            x[1] += c_a[1]
            x[2] += c_a[2]        
        x[0] /= len(A)
        x[1] /= len(A)
        x[2] /= len(A)
        y = [0, 0, 0]
        for el in B:
            c_b = image[el[0], el[1]]
            y[0] += c_b[0]
            y[1] += c_b[1]
            y[2] += c_b[2]        
        y[0] /= len(B)
        y[1] /= len(B)
        y[2] /= len(B)
        z = [0, 0, 0]
        for el in C:
            c_c = image[el[0], el[1]]
            z[0] += c_c[0]
            z[1] += c_c[1]
            z[2] += c_c[2]        
        z[0] /= len(C)
        z[1] /= len(C)
        z[2] /= len(C)
        last1 = curr1
        last2 = curr2
        last3 = curr3
        curr1 = x
        curr2 = y
        curr3 = z

    for i in range(dim_x):
        for j in range(dim_y):
            if (i, j) in A:
                image[i, j, 0] = 255
                image[i, j, 1] = 0
                image[i, j, 2] = 0
            elif (i,j) in B:
                image[i, j, 0] = 0
                image[i, j, 1] = 255
                image[i, j, 2] = 0
            elif (i,j) in C:
                image[i, j, 0] = 0
                image[i, j, 1] = 0
                image[i, j, 2] = 255

    misc.imsave('out_cat2.png', image)

if __name__ == '__main__':
    main()
```
